```python
TITANIC_TRAIN_PATH = "~/my_env/datasets/titanic/train.csv"
TITANIC_TEST_PATH = "~/my_env/datasets/titanic/test.csv"
TITANIC_GENDER_PATH = "~/my_env/datasets/titanic/gender_submission.csv"
```

```python
import pandas as pd 
train_set = pd.read_csv(TITANIC_TRAIN_PATH, delimiter = ',')
#train_set = train_set.reset_index()
train_set
```

```python
train_set.info()
```

```python
train_set = train_set.drop("Name", axis=1)
train_set = train_set.drop("Cabin", axis=1)
train_set = train_set.drop("Ticket", axis=1)
train_set = train_set.drop("PassengerId", axis=1)
```

```python
train_set
```

```python
train_set.info()
```

```python
train_set = train_set[train_set['Embarked'].notna()]
```

```python
train_set.info()
```

```python
corr_matrix = train_set.corr()
```

```python
corr_matrix["Survived"].sort_values(ascending=False)
```

```python
X_train = train_set.drop("Survived", axis=1)
y_train = train_set["Survived"].copy()
```

```python
X_train
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

num_pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy="median")),
        ('std_scaler', StandardScaler()),
    ])
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder


num_attribs = ['Pclass','Age','SibSp','Parch','Fare']
cat_attribs = ["Sex", "Embarked"]

full_pipeline = ColumnTransformer([
        ("num", num_pipeline, num_attribs),
        ("cat", OneHotEncoder(), cat_attribs),
    ])

X_train_prepared = full_pipeline.fit_transform(X_train)
```

```python
X_train_prepared.shape
```

```python
y_train.shape
```

```python
y_train = y_train.to_numpy()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold
from sklearn.base import clone

log_reg = LogisticRegression(solver="lbfgs", max_iter=5000, random_state=42)
log_reg.fit(X_train_prepared, y_train)
skfolds = StratifiedKFold(n_splits=3, random_state=42, shuffle=True)
clone_clf = clone(log_reg)
for train_index, test_index in skfolds.split(X_train_prepared, y_train): # provide train and test index such that in each fold-validation, the class 5 & class not-5 proportions of the training and test sets are the same as the entire dataset.
    X_train_folds = X_train_prepared[train_index]
    y_train_folds = y_train[train_index]
    X_test_fold = X_train_prepared[test_index]
    y_test_fold = y_train[test_index]
    
    clone_clf.fit(X_train_folds, y_train_folds)
    y_pred = clone_clf.predict(X_test_fold)
#     y_pred = y_pred.reshape(-1,1)
    n_correct = sum(y_pred == y_test_fold)
    print(n_correct / len(y_pred))
```

```python
y_pred1 = clone_clf.predict(X_train_prepared)
n_correct1 = sum(y_pred1 == y_train)
print(n_correct1 / len(y_pred1)) # So it is not overfitting...
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

param_grid = [
    # try 12 (3×4) combinations of hyperparameters
    {'n_estimators': [300,1000,3000], 'max_features': [4,5,6]},
    # then try 6 (2×3) combinations with bootstrap set as False
    #{'bootstrap': [False], 'n_estimators': [100,300], 'max_features': [4,5,6]},
  ]

forest_clf = RandomForestClassifier(random_state=42)
grid_search = GridSearchCV(forest_clf, param_grid, cv=5,
                           scoring='neg_mean_squared_error',
                           return_train_score=True)
grid_search.fit(X_train_prepared, y_train)

for_clf = grid_search.best_estimator_
skfolds = StratifiedKFold(n_splits=3, random_state=42, shuffle=True)
clone_clf = clone(for_clf)
for train_index, test_index in skfolds.split(X_train_prepared, y_train): # provide train and test index such that in each fold-validation, the class 5 & class not-5 proportions of the training and test sets are the same as the entire dataset.
    X_train_folds = X_train_prepared[train_index]
    y_train_folds = y_train[train_index]
    X_test_fold = X_train_prepared[test_index]
    y_test_fold = y_train[test_index]
    
    clone_clf.fit(X_train_folds, y_train_folds)
    y_pred = clone_clf.predict(X_test_fold)
#     y_pred = y_pred.reshape(-1,1)
    n_correct = sum(y_pred == y_test_fold)
    print(n_correct / len(y_pred))
```

```python
import numpy as np
cvres = grid_search.cv_results_
for mean_score, params in zip(cvres["mean_test_score"], cvres["params"]):
    print(np.sqrt(-mean_score), params)
```

```python
grid_search.best_params_
```

```python
from sklearn.svm import SVC

param_grid = [
    {'kernel': ['rbf'], 'C': [2,3,4], 'gamma':[0.1, 0.3, 1]},
  ]

svc_clf = SVC(random_state=42)
grid_search = GridSearchCV(svc_clf, param_grid, cv=5,
                           scoring='neg_mean_squared_error',
                           return_train_score=True)
grid_search.fit(X_train_prepared, y_train)
svc_clf = grid_search.best_estimator_
skfolds = StratifiedKFold(n_splits=3, random_state=42, shuffle=True)
clone_clf = clone(svc_clf)
for train_index, test_index in skfolds.split(X_train_prepared, y_train): # provide train and test index such that in each fold-validation, the class 5 & class not-5 proportions of the training and test sets are the same as the entire dataset.
    X_train_folds = X_train_prepared[train_index]
    y_train_folds = y_train[train_index]
    X_test_fold = X_train_prepared[test_index]
    y_test_fold = y_train[test_index]
    
    clone_clf.fit(X_train_folds, y_train_folds)
    y_pred = clone_clf.predict(X_test_fold)
#     y_pred = y_pred.reshape(-1,1)
    n_correct = sum(y_pred == y_test_fold)
    print(n_correct / len(y_pred))
```

```python
grid_search.best_params_
```

```python
from sklearn.ensemble import VotingClassifier

log_clf = LogisticRegression(solver="lbfgs", random_state=42)
rnd_clf = RandomForestClassifier(max_features=6, n_estimators=300, random_state=42)
svc_clf = SVC(gamma=0.1, probability=True, kernel = "rbf", C = 3, random_state=42)

voting_clf = VotingClassifier(
    estimators=[('lr', log_clf), ('rf', rnd_clf), ('svc', svc_clf)],
    voting='soft')
voting_clf.fit(X_train_prepared, y_train)

skfolds = StratifiedKFold(n_splits=3, random_state=42, shuffle=True)
clone_clf = clone(voting_clf)
for train_index, test_index in skfolds.split(X_train_prepared, y_train): # provide train and test index such that in each fold-validation, the class 5 & class not-5 proportions of the training and test sets are the same as the entire dataset.
    X_train_folds = X_train_prepared[train_index]
    y_train_folds = y_train[train_index]
    X_test_fold = X_train_prepared[test_index]
    y_test_fold = y_train[test_index]
    
    clone_clf.fit(X_train_folds, y_train_folds)
    y_pred = clone_clf.predict(X_test_fold)
#     y_pred = y_pred.reshape(-1,1)
    n_correct = sum(y_pred == y_test_fold)
    print(n_correct / len(y_pred))
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

y_train_pred = voting_clf.predict(X_train_prepared)
precision_score(y_train, y_train_pred)
```

```python
recall_score(y_train, y_train_pred)
```

```python
f1_score(y_train, y_train_pred) 
```
