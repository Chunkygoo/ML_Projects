```python
# We split the data in a stratified manner so that the test set has the same distribution as the training set
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(cancerDF, cancerDF[["Positive?"]]):
    strat_train_set = cancerDF.loc[train_index]
    strat_test_set = cancerDF.loc[test_index]
```

```python
# We standardize (mean=0, std=1) for the algorithms to work better
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
strat_train_set_data = scaler.fit_transform(strat_train_set_data) # fit and then transform
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold
from sklearn.base import clone

log_reg = LogisticRegression(solver='lbfgs', max_iter=5000, random_state=42)
skfolds = StratifiedKFold(n_splits=3, random_state=42, shuffle=True)
clone_clf = clone(log_reg)
for train_index, test_index in skfolds.split(strat_train_set_data, strat_train_set_labels):
    X_train_folds = strat_train_set_data[train_index]
    y_train_folds = strat_train_set_labels[train_index]
    X_test_fold = strat_train_set_data[test_index]
    y_test_fold = strat_train_set_labels[test_index]
    clone_clf.fit(X_train_folds, y_train_folds)
    y_pred = clone_clf.predict(X_test_fold)
    n_correct = sum(y_pred == y_test_fold)
    print(n_correct / len(y_pred))
```

```python
# We fit the model to the entire dataset so we can use it later if needed
log_reg.fit(strat_train_set_data, strat_train_set_labels)
```

```python
def build_model(n_hidden=3, n_neurons=30, learning_rate=3e-3, input_shape=[30]):
    model = keras.models.Sequential()
    model.add(keras.layers.InputLayer(input_shape=input_shape))
    for layer in range(n_hidden):
        model.add(keras.layers.Dense(n_neurons, activation="relu"))
    model.add(keras.layers.Dense(1, activation="sigmoid")) # activation="sigmoid" for binary outputs
    optimizer = keras.optimizers.SGD(lr=learning_rate)
    model.compile(metrics= ['acc'], loss="binary_crossentropy", optimizer=optimizer)
    return model
```

```python
keras_reg = keras.wrappers.scikit_learn.KerasClassifier(build_model)
```

```python
from scipy.stats import reciprocal
from sklearn.model_selection import RandomizedSearchCV

param_distribs = {
    "n_hidden": [0, 1, 2, 3],
    "n_neurons": np.arange(1, 100),
    "learning_rate": reciprocal(3e-4, 3e-2),
}

rnd_search_cv = RandomizedSearchCV(keras_reg, param_distribs, n_iter=10, cv=3, verbose=2)
rnd_search_cv.fit(strat_train_set_data, strat_train_set_labels, epochs=100,
                  validation_data=(strat_valid_set_data, strat_valid_set_labels),
                  callbacks=[keras.callbacks.EarlyStopping(patience=10,restore_best_weights=True)])
```
