```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(cancerDF, cancerDF[["Positive?"]]):
    strat_train_set = cancerDF.loc[train_index]
    strat_test_set = cancerDF.loc[test_index]
```

```python
strat_train_set_data = strat_train_set.drop("Positive?", axis=1) # drop labels for training set
strat_train_set_labels = strat_train_set["Positive?"].copy()
```

```python
strat_train_set_data
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
strat_train_set_data = scaler.fit_transform(strat_train_set_data)
```

```python
strat_train_set_labels = strat_train_set_labels.to_numpy()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold
from sklearn.base import clone

log_reg = LogisticRegression(solver="lbfgs", max_iter=5000, random_state=42)
log_reg.fit(strat_train_set_data, strat_train_set_labels)

skfolds = StratifiedKFold(n_splits=3, random_state=42, shuffle=True)
clone_clf = clone(log_reg)
for train_index, test_index in skfolds.split(strat_train_set_data, strat_train_set_labels): # provide train and test index such that in each fold-validation
    X_train_folds = strat_train_set_data[train_index]
    y_train_folds = strat_train_set_labels[train_index]
    X_test_fold = strat_train_set_data[test_index]
    y_test_fold = strat_train_set_labels[test_index]
    
    clone_clf.fit(X_train_folds, y_train_folds)
    y_pred = clone_clf.predict(X_test_fold)
#     y_pred = y_pred.reshape(-1,1)
    n_correct = sum(y_pred == y_test_fold)
    print(n_correct / len(y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

param_grid = [
    # try 12 (3Ã—4) combinations of hyperparameters
    {'n_estimators': [10,30,100], 'max_features': [9,10,11]},
    # then try 6 (2Ã—3) combinations with bootstrap set as False
    {'bootstrap': [False], 'n_estimators': [10,30,100], 'max_features': [9,10,11]},
  ]

forest_clf = RandomForestClassifier(random_state=42)
grid_search = GridSearchCV(forest_clf, param_grid, cv=5,
                           scoring='f1',
                           return_train_score=True)
grid_search.fit(strat_train_set_data, strat_train_set_labels)

for_clf = grid_search.best_estimator_
skfolds = StratifiedKFold(n_splits=3, random_state=42, shuffle=True)
clone_clf = clone(for_clf)
for train_index, test_index in skfolds.split(strat_train_set_data, strat_train_set_labels): # provide train and test index such that in each fold-validation
    X_train_folds = strat_train_set_data[train_index]
    y_train_folds = strat_train_set_labels[train_index]
    X_test_fold = strat_train_set_data[test_index]
    y_test_fold = strat_train_set_labels[test_index]
    
    clone_clf.fit(X_train_folds, y_train_folds)
    y_pred = clone_clf.predict(X_test_fold)
    n_correct = sum(y_pred == y_test_fold)
    print(n_correct / len(y_pred))
```

```python
cvres = grid_search.cv_results_
```

```python
cvres
```

```python
for mean_score, params in zip(cvres["mean_test_score"], cvres["params"]):
    print(np.sqrt(mean_score), params) # The higher the score the better
```

```python
from sklearn.svm import SVC
param_grid = [
    {'kernel': ['poly'], 'C': [10, 30, 100], 'gamma':[0.001, 0.003, 0.01]},
    {'kernel': ['rbf'], 'C': [10, 30, 100], 'gamma':[0.001, 0.003, 0.01]},
]

svc_clf = SVC()
grid_search = GridSearchCV(svc_clf, param_grid, cv=3,
                           scoring='f1')
                           #return_train_score=True)
grid_search.fit(strat_train_set_data, strat_train_set_labels)
svc_clf = grid_search.best_estimator_
skfolds = StratifiedKFold(n_splits=3, random_state=42, shuffle=True)
clone_clf = clone(svc_clf)
for train_index, test_index in skfolds.split(strat_train_set_data, strat_train_set_labels): # provide train and test index such that in each fold-validation, the class 5 & class not-5 proportions of the training and test sets are the same as the entire dataset.
    X_train_folds = strat_train_set_data[train_index]
    y_train_folds = strat_train_set_labels[train_index]
    X_test_fold = strat_train_set_data[test_index]
    y_test_fold = strat_train_set_labels[test_index]
    
    clone_clf.fit(X_train_folds, y_train_folds)
    y_pred = clone_clf.predict(X_test_fold)
    n_correct = sum(y_pred == y_test_fold)
    print(n_correct / len(y_pred))
```

```python
print(strat_train_set_data.shape)
print(strat_valid_set_data.shape)
print(strat_train_set_labels.shape)
print(strat_valid_set_labels.shape)
```

```python
import tensorflow as tf
from tensorflow import keras
def build_model(n_hidden=3, n_neurons=30, learning_rate=3e-3, input_shape=[30]):
    model = keras.models.Sequential()
    model.add(keras.layers.InputLayer(input_shape=input_shape))
    for layer in range(n_hidden):
        model.add(keras.layers.Dense(n_neurons, activation="relu"))
    model.add(keras.layers.Dense(1, activation="sigmoid")) # activation="sigmoid" for binary outputs
    optimizer = keras.optimizers.SGD(lr=learning_rate)
    model.compile(metrics= ['acc'], loss="binary_crossentropy", optimizer=optimizer)
    return model
```

```python
keras_reg = keras.wrappers.scikit_learn.KerasClassifier(build_model)
```

```python
from scipy.stats import reciprocal
from sklearn.model_selection import RandomizedSearchCV

param_distribs = {
    "n_hidden": [0, 1, 2, 3],
    "n_neurons": np.arange(1, 100),
    "learning_rate": reciprocal(3e-4, 3e-2),
}

rnd_search_cv = RandomizedSearchCV(keras_reg, param_distribs, n_iter=10, cv=3, verbose=2)
rnd_search_cv.fit(strat_train_set_data, strat_train_set_labels, epochs=100,
                  validation_data=(strat_valid_set_data, strat_valid_set_labels),
                  callbacks=[keras.callbacks.EarlyStopping(patience=10,restore_best_weights=True)])
```
